```python
import numpy as np 
#EChung: NumPy is an open-source numerical Python library. 
#EChung: NumPy contains a multi-dimensional array and matrix data structures. 
#EChung: It can be utilised to perform a number of mathematical operations on arrays such as trigonometric, statistical, and algebraic routines.
import matplotlib.pyplot as plt
#EChung: Rely on pyplot to automatically create and manage the figures and axes, and use pyplot functions for plotting.
#EChung: Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.
%matplotlib inline

#EChung: Constructing everything to equal to y with x on the other side of the equation
# Construct lines 
# x > 0
x = np.linspace(0, 20, 2000)
# y >= 2
y1 = (x*0) + 2
# 2y <= 25 - x
y2 = (25-x)/2.0
# 4y >= 2x - 8 
y3 = (2*x-8)/4.0
# y <= 2x - 5 
y4 = 2 * x -5

# Make plot
#EChung: ploting linear functions on the graph
#EChung: operators: beginning and the end of the equations are '$', '\geq' == >=, '\leq' == <=
plt.plot(x, y1, label=r'$y\geq2$') 
plt.plot(x, y2, label=r'$2y\leq25-x$')
plt.plot(x, y3, label=r'$4y\geq 2x - 8$')
plt.plot(x, y4, label=r'$y\leq 2x-5$')
#EChung: limits for x-axis and y-axis
plt.xlim((0, 16)) 
plt.ylim((0, 11))
#EChung: I did not yet figure out what 'label' and 'r' do
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Fill feasible region
y5 = np.minimum(y2, y4)
y6 = np.maximum(y1, y3)
plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)
#EChung: the box on the right that indicates each function
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
